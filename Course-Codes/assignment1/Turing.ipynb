# T3

## 翻译摘要

在本文中我们将讨论可能让机器表现出智能行为的方式。我们将对人类大脑的模拟作为引导原则。研究指出只有在受过合适教育的前提下人类智能的潜力才能被认知，调查主要集中于一个对机器的模拟教学上。在本文中，我们定义了未经组织化的机器，并指出婴幼儿时期的人类大脑实际上具有类似的性质。我们给出了一些例子来阐明这样的机器并讨论了利用奖惩机制来训练他们的过程。在一个例子中这一训练过程，一直进行到了这个组织类似于一个精英的水平。

接下来作者用了很大的篇幅来驳斥当时人们不看好“智能机器”的五个原因，主要是：①人们潜意识里不愿意承认“人类“会在智力上受到挑战；②宗教信仰冲突：认为试图造这样的智慧机器是一种普罗米修斯式的亵渎；③直至当时（甚至包括现在，毕竟工业革命才两百多年），“机器”在人类社会中所扮演的角色仍然很单一（可以认为这是一种成见式的原因）；④在某些应用场景下人们不愿意看到机器出错，而相对人来说，机器在某些方面纠错的灵活性太差；⑤到当时（甚至目前），机器的“智慧”更多地被看做是设计机器的人的智慧，人工智能中“人工”依然占主导地位。

## 评论观点

### 观点

在文章的11页，图灵提到，智能机器也许不应该是造一个完整的“人”，即我们不要去刻意模仿人与外界交互信息的方式（比如用麦克风代替耳朵，摄像头代替眼睛等等），而可以在一些不依赖“身体”的方面实现机器的智能。这里他举了五个例子，包括：

1. 不同的游戏，比如棋类、桥牌、扑克
2. 语言的学习能力
3. 语言的翻译能力
4. 密码学
5. 数学（这里指的应该是不依赖图表的数学，即代数）

### 正面思考

图灵的观点在今天看来也许很平常，有些方面甚至是落后的。但是站在当时时代的角度来看，他确确实实为后世人工智能的发展指了一条明路。

第一条是“游戏”，人工智能引起人们的广泛关注几乎总是从游戏开始——特别是16年横空出世的阿尔法Go。阿尔法Go背后的理论事实上早已成熟，但是它的出现真正地让资本看到深度学习实实在在的力量，只有让“外行”的资本直观地感受到技术的威力，才会有更多的资金来刺激这一行业的蓬勃发展。第二条和第三条涉及当下火热的NLP，因为语言是地球上最“人类“的东西，让机器学习和理解语言可能是所有人工智能领域的人都会感兴趣的目标。至于密码学和数学，我不清楚它们和当下火热的深度学习有何关系，不过我知道在上世纪七十年代，吴文俊先生提出的”吴方法“，在当时的几何证明领域掀起了一股机器几何的热潮，也许这离我们所期待的”机器数学能力“相差甚远，但至少可以让图灵宽慰。

### 问题所在

我觉得，就前三点而言，当下火热的深度学习算法都有自己的问题。“人工智能寒冬论”已经被一些人翻来覆去地讨论过了，我这里只想记下自己观察到的现象。

其一，关于游戏。就围棋而言，尽管最新版本的腾讯人工智能“绝艺”已经可以横扫人类棋手，但是我在观赛的过程中发现，人类顶尖棋手往往能够发现“绝艺”没有发现的妙招。我印象最深刻的是柯洁对朴廷桓的一场比赛中，人工智能给出双方的胜率是五五开的，但是柯洁发现两步妙手之后，胜率立刻倾斜到八二开。我们的“围棋上帝”并不是无所不能的：因为它不能穷举所有招数，而且它的思维方式和人类有本质的不同。所以它有两个问题：

1. 无法发现一些职业棋手都能看出来的妙手
2. 给出的所谓“好棋”让人无法理解，后续手段人类无法掌控

这两个问题结合起来就有致命的缺陷了：机器会犯错，但是人们无法理解机器的真正意图，所以即使它犯了错人也不知道或者说不理解。这个问题非常本质，如果无法规避它的话，可能会极大地限制人工智能的推广。

其二，关于语言的理解。我做过一段时间的NLP，让我比较失望的是，当时主流的方法并不“智能”，RNN在内的方法依然和我理解的“智能”有天然之别。举例来讲，“南京市长江大桥”到底啥意思，如果认为这不是内容理解而是关键词匹配，那我们到底应该在关键词匹配中如何确定“南京”、“市长”和“长江大桥”以及人名“江大桥”的优先级？这对人来说易如反掌，但是对机器学习算法来讲是一团浆糊。如果我们强行设定南京市的长江大桥为保留字，并先做一边KMP算法来滤掉这些特定字符串，那智能的意义又在哪里？如果设定对话的环境，那么每做一次迁移都会发生数据集漂移，手动去洗的话就变成“人工”而不是“智能”了。
